COMPLETAR DADOS QUE ESTÃO FALTANDO NA BASE DE DADOS

```python
# IMPORTAR AS BIBLIOTECAS
import pandas as pd
import numpy as np
```

```python
# ABRIR A BASE DE DADOS
base = pd.read_excel('/content/drive/MyDrive/dados da internet.xlsx')
x = base.iloc[:,:].values

base
```

```python
# OBS: NA BASE DE DADOS A LINHA 6 E 28 ESTÃO VAZIAS COM NOME NaN, E O OBJETIVO É COMPLETAR DE MODO ESTRATÉGICO USANDO UMA MÉDIA
```

```python
# IMPORTAR O PACOTE IMPUTER DA BIBLIOTECA SKLEARNING
from sklearn.impute import SimpleImputer
```

```python
# UTILIZAR O TRANSFORMADOR DE IMPUTAÇÃO PARA COMPLETAR VALORES AUSENTES
Imputer = SimpleImputer(missing_values=np.nan, strategy='mean')
# O IMPUTER VAI SER EM TODAS AS LINHAS E A SEGUNDA E TERCEIRA COLUNA DA MINHA VARIÁVEL X
Imputer = Imputer.fit(x[:,1:3])
# TRANSFORMAR A BASE DE VALORES NÚMEROS PARA STRING
x = Imputer.transform(x[:,1:3]).astype(str)
# INSERIR A COLUNA DE NOMES NA VARIÁVEL X, E COLOCAR ELA NA PRIMEIRA COLUNA
x = np.insert(x, 0, base.iloc[:,0].values, axis=1)

print(x)
```
